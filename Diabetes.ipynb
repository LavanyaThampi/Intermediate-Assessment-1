```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

**Univariate Analysis:**



```python
data=pd.read_csv('diabetes_model.csv')
print(data)
```

1. Plot the histogram for the dataset Diabetes

```python
plt.hist(data["Insulin"], color="green", rwidth=0.8, density=True)
plt.title("Histogram of Diabetes")
plt.xlabel("Insulin")
plt.ylabel("Density")
plt.show()
```

2. Plot the Boxplot of column Glucose for the dataset Diabetes

```python
sns.boxplot(data["Glucose"])
plt.title("Boxplot of Glucose")
plt.show()
```

**Bivariate Analysis:**


3. Plot a Scatter plot of Blood pressure and BMI for the dataset Diabetes.

```python
plt.figure(figsize=(5, 3.5))
plt.scatter(data["BloodPressure"], data["BMI"], c="k", marker="*")
plt.title("BMI VS BloodPressure")
plt.xlabel("BloodPressure")
plt.ylabel("BMI")
plt.show()
```

4. Correlation matrix for the given dataset Diabetes.

```python
with open("diabetes_model.csv") as file:
    df = pd.read_csv(file)
num_columns = df.select_dtypes(include='number').columns
corr_matrix = df[num_columns].corr()
print("Correlation matrix:")
print(corr_matrix)
```

**Multivariate Analysis:**


5. Plot the pair plot for the dataset Diabetes

```python
sns.pairplot(data, hue="Outcome", x_vars=['SkinThickness'])
plt.show()
```

6. 3D Plot for the Dataset Diabetes

```python
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.set_xlabel('Pregnancies')
ax.set_ylabel('Glucose')
ax.set_zlabel('Age')
ax.set_title('3D Plot of Pregnancies, Glucose, and Age')
ax.scatter(data['Pregnancies'], data['Glucose'], data['Age'], c=data['Outcome'], cmap='viridis')
```
